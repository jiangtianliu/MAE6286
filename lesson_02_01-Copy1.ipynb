```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = 41
nt = 18
dt = 0.02
dx =2/(nx-1)
x = numpy.linspace(0, 2, nx)
u = numpy.ones(nx)-1
for i in range(nx):
    if x[i]>=0.9 and x[i]<=1.0:
        u[i] = 10*(x[i]-0.9)
    if x[i]>=1.0 and x[i]<=1.1:
        u[i] = 10*(1.1-x[i])
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0, 1.5);
```

```python
for n in range(1, nt):  
    un = u.copy() 
    u[2:] = un[2:]-1*dt/(2*dx)*(un[2:]-un[0:-2]) 
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(-0.5, 0.5);
```

```python
u[0] = 0.0# Boundary condition
for n in range(1, nt):
    un = u.copy()
    u[1:] = un[1:]-un[1:]*dt/dx*(un[1:]-un[0:-1])
```

```python
pyplot.plot(x, u, color='#003366', ls='--', lw=3)
pyplot.ylim(0, 2.5);
```
